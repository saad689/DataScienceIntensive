I have had very poor experiences flying with United Airlines departing out of O'Hare Airport. I will now conduct a hypothesis test to determine if United Airlines performance in terms of delayed departures is worse than the population at O'Hare.

The hypothesis test will be set up as follows:

    * H0: p_ua - p_notUA = 0 i.e. United Airlines is not worse than the average airline out of O'Hare
    * HA: p_ua > p_notUA i.e. United Airlines is 
    

```python
#Calculate estimate of p_UA and p_~UA
s_tot_del_UA = len(flt_df[(flt_df.DEP_DELAY >= 15) & (flt_df.UNIQUE_CARRIER == 'UA')])
s_tot_UA = len(flt_df[(flt_df.UNIQUE_CARRIER == 'UA')])
p_UA = float(s_tot_del_UA)/s_tot_UA
s_tot_del_nUA = len(flt_df[(flt_df.DEP_DELAY >= 15) & (flt_df.UNIQUE_CARRIER != 'UA')])
s_tot_nUA = len(flt_df[(flt_df.UNIQUE_CARRIER != 'UA')])
p_notUA = float(s_tot_del_nUA)/s_tot_nUA
p_pool = float(s_tot_del_UA + s_tot_del_nUA)/(s_tot_UA + s_tot_nUA)
```

```python
#Calculate the Standard Error and then the Z-Statistic
SE = np.sqrt((p_pool*(1-p_pool))/s_tot_UA + (p_pool*(1-p_pool))/s_tot_nUA)
z_stat = (p_UA - p_notUA)/SE
p_value = 1 - scipy.stats.norm.cdf(z_stat)
print 'Hypothesis Test Results'
print 'z_stat = ', z_stat, '\np-value = ', p_value, ', sig_level = 0.05'
```

From the result above at a significance level of 0.05 we can reject the null hypothesis and conclude from the data above that United Airlines performs worse than the average of the rest of the population of airlines at O'Hare international airport. 

###How much does the airline indicate the possibility of delay?
The next question I would like to answer is how much does the airline affect the possibility of a delayed flight? To determine that, I will conduct a chi square test of independence.

```python
df_del_flt = pd.DataFrame([s_tot_del_flt, s_tot_ot_flt, s_tot_del_flt + s_tot_ot_flt], index=['Delayed', 'OnTime', 'Total'])

df_del_flt.transpose()
```

The last test that will be conducted as part of this data story is determining whether the fact that a flight is delayed or not is dependent on the airline. To do this, we will conduct a Chir-Square Independence Test."

```python
#Conduct Chi Square Test
s_tot_del = flt_df[flt_df.DEP_DELAY >= 15].groupby('UNIQUE_CARRIER').size()
s_tot_ot = flt_df[flt_df.DEP_DELAY < 15].groupby('UNIQUE_CARRIER').size()
s_tot = flt_df.groupby('UNIQUE_CARRIER').size()
p_ind = float(s_tot_del.sum())/s_tot.sum()

s_tot_del_exp = s_tot*p_ind
chi_sq, p_value = scipy.stats.chisquare(s_tot_del, f_exp=s_tot_del_exp, ddof=10)
p_value, chi_sq
```

From the result above it appears that the p_value is lower than the significance value of 0.05. We can reject the null hypothesis and say that there is a relationship between airline and proportion of flights delayed.

###How much worse or better than other airports is O'Hare in terms of delayed departures?

```python
s_tot_del = flt_df[flt_df.DEP_DELAY >= 15].groupby(['MONTH', 'DAY_OF_MONTH', 'DAY_OF_WEEK']).size()
s_tot = flt_df.groupby(['MONTH', 'DAY_OF_MONTH', 'DAY_OF_WEEK']).size()
pct_tot_del = s_tot_del/s_tot
pct_tot_del.plot(title='Histogram of Delay Pct Per Day', kind='hist')
```

A hypothesis test will be conducted to determine if O'Hare Airport is significantly worse than the population in performance in terms of delayed departures. The known population proportion of percentage of delayed flights in 2015 is 19.62% as provided by the Bureau of Transportation Statistics: http://www.transtats.bts.gov/HomeDrillChart.asp?URL_SelectMonth=8&URL_SelectYear=2015

The hypothesis test will be set up as follows:

    * H0: p_delay = 0.1962, i.e. O'Hare's proportion of delayed flights is equal to the population delay
    * HA: p_delay > 0.1962, i.e. O'Hare's proportion of delayed flights is greater than the population delay
    
Next, the conditions for the Central Limit Theorem are checked:

    1) n = 243 which is less than 10% of proportions of delayed flights each day at every airport
    2) The sample looks skewed to the right: n*p = 243*0.1962 > 10 and n*(1-p) = 243*(1-0.1962) > 10

Appears that the conditions for the central limit theorem are met

```python
#Conduct hypothesis test by calculating the z test statistic
n, p = len(pct_tot_del), 0.1962
SE = np.sqrt(p*(1-p)/n)
z_stat = (pct_tot_del.mean() - p)/SE
p_value = 1 - scipy.stats.norm.cdf(z_stat)
print 'Hypothesis Test Results'
print 'n = ', n, ', p = ', p, ', SE = ', SE, ', z_stat = ', z_stat, '\np-value = ', p_value, ', sig_level = 0.05'
```

Based on the hypothesis test conducted above, at a significance level of 0.05 the null hypothesis can be rejected and we can conclude that O'Hare airport performs worse in terms of pcercent of delayed flight departures than the average airport.
